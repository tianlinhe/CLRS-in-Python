**10.1-4
Rewrite ENQUEUE and DEQUEUE to detect underflow and overflow of a queue**.

* Queue underflow: when we attempt to dequeue an element from an empty queue
* Queue overflow: When we attempt to enqueue an element to a full queue
Therefore, we added the function `queue_full` to check if a queue is full, and 
1. checks `queue_empty` before `dequeue`
2. checks `queue_full` before `enqueue`

```python
class Queue:
    def __init__(self,n): # build a stack with an n-element empty array
        self.array=[None]*n
        self.head=0 # initial head=0
        self.tail=0 # initial tail=0
        self.length=n # length=n, arbitary and unchanged
        
    def queue_empty(self): # check if the queue is empty
        if self.head==self.tail and self.array[self.head] is None:
            return True
        else:
            return False
        
    def queue_full(self): # check is the queue is full
        if self.head==self.tail and self.array[self.head] is not None:
            return True
        else:
            False
        
    def enqueue(self,x):
        if self.queue_full() is True: # if queue is full
           # print ('queue overflows')
            return
        else:
            self.array[self.tail]=x 
            if self.tail==self.length-1: # if tail reaches the largest index (i.e. n-1) of the array 
                self.tail=0 # x is inserted on index=0 
            else:           
                self.tail+=1 # x is inserted on the right, move tail 1 step towards right

    def dequeue(self):
        if self.queue_empty() is True: # if queue is empty
           # print ('queue underflows')
            return
        else:
            x=self.array[self.head] # item to be dequeue is located at head
            self.array[self.head]=None # dequeue
            if self.head==self.length-1: # if head  reaches the largest index (i.e. n-1) of the array
                self.head=0 # re-position head from the LHS
            else:
                self.head+=1 # move head 1 step towards right
            return x      
```

**10.1-5
Whereas a stack allows insertion and deletion of elements at only one end, and a queue allows insertion at one end and deletion a the other ends, a deque (double-ended queue) allows insertion and deletion at both ends. Write four O(1)-time procedures to insert elements into and delete elements from both ends of a deque implemented by an array.**

We constructed `Deque` on the basis of `Queue`. Notice that:
1. `insert_from_right` is the equivalence to ENQUEUE
2. `delete_from_left` is the equivalence of DEQUEUE

```python
class Deque:
    def __init__(self,n):
        self.array=[None]*n # build a deque with an n-element empty array
        self.head=0
        self.tail=0
        self.length=n
    def queue_empty(self): # check if the queue is empty
        if self.head==self.tail and self.array[self.head] is None:
            return True
        else:
            return False
        
    def queue_full(self): # check is the queue is full
        if self.head==self.tail and self.array[self.head] is not None:
            return True
        else:
            False
        
    def insert_from_right(self,x): # equivalence to ENQUEUE
        if self.queue_full() is True: # if queue is full
            print ('queue overflows')
            return
        else:
            self.array[self.tail]=x 
            if self.tail==self.length-1: # if tail reaches the rightmost index (i.e. n-1) of the array 
                self.tail=0 # x is inserted on index=0 
            else:           
                self.tail+=1 # x is inserted on the right, move tail 1 step towards right
                
    def insert_from_left(self,x): # modified from ENQUEUE
        if self.queue_full() is True: 
            print ('queue overflows')
            return
        else:
            if self.head==0: # if head reaches the leftmost index (i.e. 0) of the array 
                self.head=self.length-1 # re-position head to rightmost index 
            else:           
                self.head-=1 #  move head 1 step towards left
            self.array[self.head]=x

    def delete_from_left(self): # equivalence to DEQUEUE
        if self.queue_empty() is True: # if queue is empty
            print ('queue underflows')
            return
        else:
            x=self.array[self.head] # item to be dequeue is located at head
            self.array[self.head]=None # set x as none
            if self.head==self.length-1: # if head  reaches the rightmost index (i.e. n-1) of the array
                self.head=0 # re-position head to the leftmost index
            else:
                self.head+=1 # move head 1 step towards right
            return x
    def delete_from_right(self): 
        if self.queue_empty() is True: 
            print ('queue underflows')
            return
        else:
            if self.tail==0: 
                self.tail=self.length-1 
            else:
                self.tail-=1 
            x=self.array[self.tail]
            self.array[self.tail]=None #set x as none
            return x
```

**10.1-6
Show how to implement a queue using two stacks. Analyze the running time of the queue operations.**

* ENQUEUE is the equivalence of PUSH, therefore it takes O(1) time
* DEQUEUE takes O(n) time in the worst case:
    1. Pop item from stack 1 and push it to stack 2
    2. Repeat step 1 until stack 1 is empty
    3. Pop item from stack 2 and return the item
**Notice that you will need to first run the block `class Stack` that we constructed in 10.1_Stacks_and_queues.ipynb**

```python
class Queue_with2stacks():
    def __init__(self,n):
        self.s1=Stack(n=int(n/2))
        self.s2=Stack(n=int(n/2))
           
    def enqueue(self,x): #enqueue is the same as push
        self.s1.push(x) # push x into s1
    
    def dequeue(self): #pop an element from s2
        
        if self.s2.stack_empty() is True:
        """ we pop it from s1, in the order from right to left
             we push it to s2, in the order from left to right"""
            self.s2.push(self.s1.pop())
                
        self.s2.pop() # the rightmost from s2 is pop
        return(self.s2.array[self.s2.top+1]) #return the poped item
```

**10.1-7
Show how to implement a stack using two queues. Analyze the running time of the stack operations.**

- PUSH is equivalent to ENQUEUE, therefore it takes O(1) time
- POP that takes O(n) time:
    1. Dequeue items from one of the queues and place it in the other
    2. Repeat step 1 until we have one item left
    3. Return this item as POPed
    4. Dequeue this item from its queue
The POP procedures is graphically illustrated in figure 1: `![figure 1: POP](img/10.1-6.png)`
    
**Notice that you will need to first run the block `class Queue` that we constructed in Exc.10.1-4 from 10.1_Stacks_and_queues_Exercise.ipynb**

```python
class Stack_with2queues:
    def __init__(self,n):
        self.q1=Queue(n=int(n/2))
        self.q2=Queue(n=int(n/2))
        
    def push(self,x): # push is equivalence to enqueue
        self.q1.enqueue(x)
      
        
    def pop(self):
        """ if q1 is not empty:
            dequeue from q1 until only one item is left (and return the item as poped)
            at the same time enqueue q2"""
        if self.q1.queue_empty() is False: 
            
            #execute the while loop until only one item from q1 is left
            while self.q1.tail-self.q1.head != 1 and self.q1.tail-self.q1.head!= 1-self.q1.length:
             
                self.q2.enqueue(self.q1.dequeue())
                
            x=self.q1.array[self.q1.head] #return the item as x
            self.q1.dequeue() # finally, empty q1 completely
         
            return x 
        
        """ if q2 is not empty, similar steps:
            dequeue from q2 until only one item is left (and return the item as poped)
            at the same time enqueue q1"""    
        else: 
            
            while self.q2.tail-self.q2.head != 1 and self.q2.tail-self.q2.head!= 1-self.q2.length:
                self.q1.enqueue(self.q2.dequeue())
                
            x=self.q2.array[self.q2.head]
            self.q2.dequeue()
            
            return x 
```
